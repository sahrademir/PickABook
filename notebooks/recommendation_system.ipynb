```python
books_dev = books.sample(
    n=100000,
    random_state=42
).reset_index(drop=True)
```

```python
books_dev.shape
```

```python
tfidf = TfidfVectorizer(stop_words="english")

tfidf_matrix = tfidf.fit_transform(
    books_dev["Description"]
)
```

```python
tfidf_matrix.shape
```

```python
indices = pd.Series(
    books_dev.index,
    index=books_dev["Name"]
).drop_duplicates()
```

```python
indices.head()
```

```python
books_dev["Name"].sample(10, random_state=42)
```

```python
indices["Toad"]
```

```python
book_index = indices["Toad"]  
```

```python
book_vector = tfidf_matrix[book_index]
```

```python
similarities = cosine_similarity(
    book_vector,
    tfidf_matrix
)
```

```python
similarities.shape
```

```python
similarities = similarities.flatten()
```

```python
similarities.shape
```

```python
import numpy as np

top_indices = np.argsort(similarities)[::-1]
```

```python
top_indices[1:11] #The first index is the book itself; the similarity rate is 100%. That's why 1:11
```

```python
books_dev.iloc[top_indices[1:11]][
    ["Name", "Authors"]
]
```

```python
books_dev["Features"] = (
    books_dev["Name"].fillna("") + " " +
    books_dev["Authors"].fillna("") + " " +
    books_dev["Description"].fillna("")
)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
books_dev["Features"] = books_dev["Features"].apply(clean_text)
```

```python
def recommend(book_name, top_n=10):
    
    if book_name not in indices:
        return "Book not found."

    idx = indices[book_name]

    similarity_scores = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    similar_indices = similarity_scores.argsort()[::-1]

    similar_indices = similar_indices[1:top_n+1]

    recommendations = books_dev.iloc[similar_indices][
        ["Name", "Authors", "Rating"]
    ].copy()

    recommendations["Similarity"] = similarity_scores[similar_indices]

    return recommendations
```

```python
recommend("Toad")
```
